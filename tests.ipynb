```python
from target_encoder import TargetEncoder
from sparse_onehot_encoder import RobustOneHotSparse
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
iris_data = datasets.load_iris()
data = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
data['cat_col'] = np.random.randint(100, size = data.shape[0])
data['cat_col'] = data['cat_col'].astype('category')
data['target'] = pd.Series(iris_data.target)
data.head()
```

```python
oh_enc = RobustOneHotSparse()
t_enc = TargetEncoder(type = 'categorical')
```

```python
oh_enc.fit(data, columns = 'cat_col')
t_enc.fit(data, columns = 'cat_col', target = 'target')
```

```python
t_enc.transform(data)
```
